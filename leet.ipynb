```python
import spacy
from time import time

textao = text.join(" He said no more, but slowly waving a benediction.")

def extract_token_pos(text, nlp):
    doc = nlp(text)
    
    verb = [] # Verbo (should provide, enable, etc.)
    aux = [] # Auxiliar (should, must, can, is, are, etc.)
    noun = [] # Substantivo (system, user, report, etc.)
    propn = [] # Substantivo próprio (Nome da empresa, nome de funcionário, etc.)
    pron = [] # Pronome (user, it, they, he, she, etc.)
    adj = [] # Adjetivo (detailed, new, old, etc.)
    adv = [] # Advérbio (automatically, only, quickly, slowly, etc.)
    sconj = [] # Conjunção subordinada (so, that, when, if, because, etc.)
    part = [] # Particípio (to provide, to allow, should be, should have, etc.)
    org = [] # Organização (Nome da empresa, nome de funcionário, etc.)
    
    for token in doc:
        token_text = token.text
        pos = token.pos_

        if pos == "VERB":
            verb.append(token_text)
        elif pos == "AUX":
            aux.append(token_text)
        elif pos == "NOUN":
            noun.append(token_text)
        elif pos == "PROPN":
            propn.append(token_text)
        elif pos == "PRON":
            pron.append(token_text)
        elif pos == "ADJ":
            adj.append(token_text)
        elif pos == "ADV":
            adv.append(token_text)
        elif pos == "SCONJ":
            sconj.append(token_text)
        elif pos == "PART":
            part.append(token_text)
        elif pos == "ORG":
            org.append(token_text)
    return verb, aux, noun, propn, pron, adj, adv, sconj, part, org

def extract_token_pos_opt(text, nlp):
    doc = nlp(text)
    dict_to_pos_tag = {"VERB": [], "AUX": [], "NOUN": [], "PROPN": [], "PRON": [], "ADJ": [], "ADV": [], "SCONJ": [], "PART": [], "ORG": []}
    #verb: Verbo (should provide, enable, etc.)
    #aux: Auxiliar (should, must, can, is, are, etc.)
    #noun: Substantivo (system, user, report, etc.)
    #propn: Substantivo próprio (Nome da empresa, nome de funcionário, etc.)
    #pron: Pronome (user, it, they, he, she, etc.)
    #adj: Adjetivo (detailed, new, old, etc.)
    #adv: Advérbio (automatically, only, quickly, slowly, etc.)
    #sconj: Conjunção subordinada (so, that, when, if, because, etc.)
    #part: Particípio (to provide, to allow, should be, should have, etc.)
    #org: Organização (Nome da empresa, nome de funcionário, etc.)
    
    for token in doc:
        if token.pos_ in dict_to_pos_tag.keys():
            dict_to_pos_tag[token.pos_].append(token.text)

    return dict_to_pos_tag


a = time()
extract_token_pos(textao, spacy.load("en_core_web_sm"))
b = time()
extract_token_pos_opt(textao, spacy.load("en_core_web_sm"))
c = time()
print(f"b-a: {b-a}")
print(f"c-b: {c-b}")
```

```python
import matplotlib.pyplot as plt
import spacy
import time
import numpy as np

# Load SpaCy model
nlp = spacy.load("en_core_web_sm")

# Define both functions
def extract_token_pos(text, nlp):
    doc = nlp(text)
    verb, aux, noun, propn, pron, adj, adv, sconj, part, org = [], [], [], [], [], [], [], [], [], []
    
    for token in doc:
        token_text = token.text
        pos = token.pos_

        if pos == "VERB":
            verb.append(token_text)
        elif pos == "AUX":
            aux.append(token_text)
        elif pos == "NOUN":
            noun.append(token_text)
        elif pos == "PROPN":
            propn.append(token_text)
        elif pos == "PRON":
            pron.append(token_text)
        elif pos == "ADJ":
            adj.append(token_text)
        elif pos == "ADV":
            adv.append(token_text)
        elif pos == "SCONJ":
            sconj.append(token_text)
        elif pos == "PART":
            part.append(token_text)
        elif pos == "ORG":
            org.append(token_text)
    return verb, aux, noun, propn, pron, adj, adv, sconj, part, org

def extract_token_pos_opt(text, nlp):
    doc = nlp(text)
    dict_to_pos_tag = {"VERB": [], "AUX": [], "NOUN": [], "PROPN": [], "PRON": [], "ADJ": [], "ADV": [], "SCONJ": [], "PART": [], "ORG": []}
    
    for token in doc:
        try:
            dict_to_pos_tag[token.pos_].append(token.text)
        except:
            pass

    return dict_to_pos_tag

# Generate texts of increasing length
base_text = "As a user, I want to quickly access the system so that I can generate reports."
scales = [10, 50, 100, 200, 400, 800, 1600, 3200, 6400, 10000]
texts = [base_text * scale for scale in scales]

# Measure time for each
times_original = []
times_opt = []

for text in texts:
    # Original
    start = time.time()
    extract_token_pos(text, nlp)
    times_original.append(time.time() - start)
    
    # Optimized
    start = time.time()
    extract_token_pos_opt(text, nlp)
    times_opt.append(time.time() - start)

# Plot results
plt.figure(figsize=(10, 6))
plt.plot(scales, times_original, label="extract_token_pos (original)", marker="o")
plt.plot(scales, times_opt, label="extract_token_pos_opt (optimized)", marker="o")
plt.xlabel("Text Size (repeats of base sentence)")
plt.ylabel("Execution Time (seconds)")
plt.title("Comparison of POS Tagging Functions")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
